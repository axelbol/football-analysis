```python
# delete columns reassigning to the DF

df_possession_clean = df.drop(['Unnamed: 1_level_0_#', 'Unnamed: 2_level_0_Nation', 'Unnamed: 3_level_0_Pos', 'Unnamed: 4_level_0_Age', 'Unnamed: 5_level_0_Min', 'Touches_Live', 'Take-Ons_Att', 'Take-Ons_Succ', 'Take-Ons_Succ%', 'Take-Ons_Tkld', 'Take-Ons_Tkld%', 'Carries_Carries', 'Carries_TotDist', 'Carries_PrgDist', 'Carries_PrgC', 'Carries_1/3', 'Carries_CPA', 'Carries_Mis', 'Carries_Dis', 'Receiving_Rec', 'Receiving_PrgR'], axis=1)
```

```python
# delete columns without reassigning to the DF

df.drop(['Unnamed: 1_level_0_#', 'Unnamed: 2_level_0_Nation', 'Unnamed: 3_level_0_Pos', 'Unnamed: 4_level_0_Age', 'Unnamed: 5_level_0_Min', 'Touches_Live', 'Take-Ons_Att', 'Take-Ons_Succ', 'Take-Ons_Succ%', 'Take-Ons_Tkld', 'Take-Ons_Tkld%', 'Carries_Carries', 'Carries_TotDist', 'Carries_PrgDist', 'Carries_PrgC', 'Carries_1/3', 'Carries_CPA', 'Carries_Mis', 'Carries_Dis', 'Receiving_Rec', 'Receiving_PrgR'], axis=1, inplace=True)
```

```python
df_possession_clean
```

```python
# rename column labels

df_possession_clean.rename(columns={'Unnamed: 0_level_0_Player': 'Player', 'Touches_Touches': 'Touches', 'Touches_Def Pen': 'DefPen', 'Touches_Def 3rd': 'Def3rd', 'Touches_Mid 3rd': 'Mid_3rd', 'Touches_Att 3rd': 'Att_3rd', 'Touches_Att Pen': 'AttPen'}, inplace=True)
df_possession_clean
```

```python
# sort data by touches

df_possession_clean = df_possession_clean.sort_values(by=['Touches'], ascending=False)
df_possession_clean
```

```python
# delete row with player's count

df_possession_clean.drop(16, inplace=True)
df_possession_clean
```

```python
# delete touches column

df_possession_clean = df_possession_clean.drop('Touches', axis=1)
df_possession_clean
```

```python
# transpose the DF

df_possession_clean = df_possession_clean.T
df_possession_clean
```

```python
# move column names

df_possession_clean.columns = df_possession_clean.iloc[0]
df_possession_clean
```

```python
# drop player row

df_possession_clean = df_possession_clean.drop('Player')
df_possession_clean
```

```python
players_list = df_possession_clean.columns.values.tolist()
players_list
```

```python
# remove player

df_possession_clean = pd.DataFrame(df_possession_clean, columns=players_list)
df_possession_clean
```

```python
# df_possession_clean['Hal Hershfelt'].dtype
df_possession_clean.info()
```

```python
# List the columns you want to convert to int

df_possession_clean['Hal Hershfelt'] = df_possession_clean['Hal Hershfelt'].astype(int)
```

```python
# convert all DF to int

df_possession_clean = df_possession_clean.astype(int)
```
